```python
import pandas as pd

# Read the dataset
df = pd.read_csv('2024Nov_HT_Phase2.csv', encoding='latin-1')  # Replace with your actual file path

# Extract the unique values from the specific column (e.g., 'column_name')
unique_values = df['comorbidity'].unique()

# Print the unique values
for value in unique_values:
    print(value)

```

```python
print(df['comorbidity'].to_list())
```

```python
# Get the count of unique values in the specific column (e.g., 'column_name')
value_counts = df['comorbidity'].value_counts()

# Print the unique values and their counts
for value, count in value_counts.items():
    print(f"Unique Value: {value}, Count: {count}")
```

```python
all_terms = df['comorbidity'].str.split(',').explode().str.strip()

# Get the count of unique terms
term_counts = all_terms.value_counts()

# Print the unique terms and their counts
for term, count in term_counts.items():
    print(f"Unique Term: {term}, Count: {count}")
```
