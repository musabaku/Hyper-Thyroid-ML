```python
# Define the pattern to remove
import re
pattern = r'\s*(gr|mg|gram)/gün\s*'

# Function to remove specified patterns from a string
def remove_patterns(text):
    if isinstance(text, str):
        return re.sub(pattern, '', text)
    else:
        return text

# Columns to clean
columns_to_clean = ['urinary_calcium_24h', 'urinary_creatinine_24h', 'urinary_phosphorus_24h']

# Apply the function to specified columns
for col in columns_to_clean:
    df[col] = df[col].apply(remove_patterns)

print(df)
```
