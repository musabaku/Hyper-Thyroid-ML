```python
import re

df['serum_creatinine'] = df['serum_creatinine'].astype(str).str.replace(r'\.{2,}', '.', regex=True)

print(df['Triglycerides'].head(50))

def clean_triglycerides(x):
    if isinstance(x, str):
        x = x.strip()
        if x == "":
            return ""
        # Split the string by semicolon
        parts = x.split(';')
        values = []
        for part in parts:
            # Remove non-numeric characters, then strip whitespace
            cleaned = re.sub(r'[^0-9.-]', '', part).strip()
            if cleaned:
                try:
                    values.append(float(cleaned))
                except ValueError:
                    pass
        # Return the highest value found, or an empty string if no valid number is found
        return max(values) if values else ""
    return x

df['Triglycerides'] = df['Triglycerides'].apply(clean_triglycerides)

print(df['Triglycerides'].head(50))
```

# Clean messy data format

```python
import re

def robust_convert(x):
    # Ensure we work with a string
    if not isinstance(x, str):
        return x
    # Trim whitespace; if the string is empty, leave it empty
    x = x.strip()
    if x == "":
        return ""
    # Split on semicolon to handle multiple values in one cell
    parts = x.split(';')
    values = []
    for part in parts:
        part = part.strip()
        # Replace commas with dots and remove square brackets if present
        part = part.replace(',', '.').strip('[]')
        # Use regex to extract numeric parts (including negatives and decimals)
        found_numbers = re.findall(r'-?\d+(?:\.\d+)?', part)
        for num in found_numbers:
            try:
                values.append(float(num))
            except Exception:
                pass
    # If no valid numbers were found, leave it empty; otherwise, return the maximum value
    return max(values) if values else ""

# List of columns to convert
columns_to_convert = [
    'bmi', 'magnesium', 'alp', 'albumin', 'ionized_calcium', 'serum_creatinine',
    'phosphorus', 'urinary_calcium_24h', 'urinary_creatinine_24h', 'Triglycerides',
    'ldl_cholesterol', 'ast', 'ggt', 'bun', 'chlorine', 'serum_25_hydroxy_vitamin_d',
    'pth', 'l1_t_score', 'l4_t_score', 'femoral_neck_z_score', 'femoral_neck_t_score',
    'radius_z_score', 'pf_index'
]

# Apply the robust conversion function to each column
for col in columns_to_convert:
    df[col] = df[col].astype(str).apply(robust_convert)

# Optionally, print some unique values to check the conversion
for col in columns_to_convert:
    unique_vals = df[col].dropna().unique()
    print(f"Column: {col}")
    print(unique_vals)
    print("-" * 30)
```
